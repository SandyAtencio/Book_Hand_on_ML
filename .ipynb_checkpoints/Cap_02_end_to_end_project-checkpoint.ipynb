### Obtener los datos

```python
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join('datasets','housing')
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def obtener_data(housing_url = HOUSING_URL, housing_path = HOUSING_PATH):
    """
    permite obtener los datos desde otras fuentes de manera dinamica
    al llamar esta funcion, se creara un directorio: datasets/housing en el espacio de trabajo
    luego descarga el archivo housing.tgz y extrae el housing.csv.
    """
    #crear directorio recursivamente(va a crear un directorio de forma datasets/housing)
    os.makedirs(housing_path, exist_ok=True)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    #permite descargar los archivos remoto a local
    urllib.request.urlretrieve(housing_url, tgz_path)
    #permite abrir el directorio que ya esta creado
    housing_tgz = tarfile.open(tgz_path)
    # descomprimir el archivo y escribir los archivos en el sistema de archivos
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
def cargar_data(housing_path= HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
```

### Transformedores personalizados

se crea un transfromador personalizado para agregar atributos adicionales

```python
from sklearn.base import BaseEstimator, TransformerMixin

# indice de columna
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True): # sin *args o **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
data_extra_attribs = attr_adder.transform(data.values)
```

```python
col_names = "total_rooms", "total_bedrooms", "population", "households"
rooms_ix, bedrooms_ix, population_ix, households_ix = [
    data.columns.get_loc(c) for c in col_names] 
```

### Pipelines de transformacion

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('attribs_adder', CombinedAttributesAdder()),
        ('std_scaler', StandardScaler()),
])

data_num_tr = num_pipeline.fit_transform(data_num)
```

```python
data_num_tr
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(data_num)
cat_attribs = ['ocean_proximity']

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])
```

```python
data_prepared = full_pipeline.fit_transform(data)
```

```python
data_prepared
```

### Seleccionar y entrenar un modelo

#### LinearRegression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(data_prepared, data_labels)
```

```python
#se seleccionan las primeras 5 filas para observar algunas prediciones y sus comportamiento
some_data = data.iloc[:5]
some_labels = data_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
values_predict = lin_reg.predict(some_data_prepared)

comparison_df = pd.DataFrame(data={"Original Values": some_labels, "Predicted Values": values_predict})

comparison_df['Differences'] = comparison_df["Original Values"] - comparison_df["Predicted Values"]
comparison_df
#print("prediciones: ", lin_reg.predict(some_data_prepared))
#compramos con los valores reales
```

a simple vista podemos ver que las prediciones no son exactamente precisas, por ejemplo la primera prediccion tiene un error de cerca del 40%.

entonces se procede a medir el RMSE de este modelo de regresión en todo el conjunto de train

```python
from sklearn.metrics import mean_squared_error

data_predictions = lin_reg.predict(data_prepared)
lin_mse = mean_squared_error(data_labels, data_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

aun sigue siendo un mal puntaje, ya que la mayoria del median_housing_value oscilan entre $120.000 y $265.000 por lo que un error de predicción típico de $68.628 no es muy satisfactorio, entonces se puede decir que este modelo esta subajustado en el conjunto de train

#### DecisionTreeRegressor

hay varias opciones para tratar el subajuste de nuestro modelo, en este caso tomamos la opcion de intentar crear un modelo mas complejo, usaremos un **DecisionTreeRegressor**

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(data_prepared, data_labels)
```

una vez entrenado el modelo, procedemos a evaluarlo en el conjunto de train

```python
data_predictions = tree_reg.predict(data_prepared)
tree_mse = mean_squared_error(data_labels, data_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

vemos que no hay error, es probable que el modelo haya sobreajustado los datos, y para ello debemos tomar una parte del conjunto de train y tomarlo como un conjunto de validacion.

### Mejor evaluación mediante validación cruzada (cross-validation)

con base en lo anterior se entrenara el modelo con el conjunto train mas pequeño y luego se evalua con el conjunto de validacion. para ello utilizaremos K-fold cross-validation lo cual divide aleatoriamente el conjunto de train en K subconjuntos distintos llamados folds, luego entrena y evalua el modelo K veces, elije un folds(pliegue) diferente cada vez para la evaluación y entrena con los otros k-1 folds, finalmente el resultado es una matriz que contiene los k puntuaciones de la evaluacion



```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, data_prepared, data_labels, 
                        scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_score(scores):
    print("Scores: ", scores)
    print("Mean: ", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
display_score(tree_rmse_scores)
```

```python
lin_scores = cross_val_score(lin_reg, data_prepared, data_labels, 
                        scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
```

```python
display_score(lin_rmse_scores)
```

se observa que el model Decision Tree tiene peor rendimiento que el Modelo de regresion lineal.
Vamos a probar un ultimo modelo, Random Firest Regresor, funciona entrenado muchos arboles de decision en subconjuntos aleatorios de la caracteristicas y luego promedia sus prediciones.

```python
data_predictions = forest_reg.predict(data_prepared)
forest_mse = mean_squared_error(data_labels, data_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
forest_scores = cross_val_score(forest_reg, data_prepared, data_labels, 
                        scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
```

```python
display_score(forest_rmse_scores)
```

el puntaje de este modelo en el conjunto de train sigue siendo mucho mas bajo que en el conjunto de validacion, esto significa que el modelo sigue teniendo sobreajuste en el conjunto de train

### Ajustar el modelo

### Grid Search

permite encontrar la mejor combinacion de hiperparametros


```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    # prueba 12 combinaciones de hyperparametros (3x4)
    {'n_estimators': [3, 10, 30], 'max_features':[2,4,6,8]},
    # prueba 6 combinaciones de hyperparametros pero con bootstrap en False (2x3)
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2,3,4]},
]

forest_reg = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(forest_reg, param_grid, cv = 5,
                          scoring = 'neg_mean_squared_error',
                          return_train_score=True)

grid_search.fit(data_prepared, data_labels)
```

Dos fromas de saber cual es la mejor combinacion de hyperparametros encontrada

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

aqui se puede observar el puntaje por cada commbinacion de hyperparametro

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

se obtiene que la mejor combinacion es 8 en max_features y 30 en n_estimators, el RMSE para esta combinacion es de 49,898 se puede decir que es mucho mejor que el score que habiamos encontrados con los valores de hyperparametros por defecto (el cual fue 50,435)

```python
feature_importances = grid_search.best_estimator_.feature_importances_
feature_importances
```

```python
extra_attribs = ['rooms_per_hold', 'pop_per_hhold', 'bedrooms_per_room']
cat_encoder = full_pipeline.named_transformers_['cat']
cat_one_hot_attribs = list(cat_encoder.categories_[0])
attributes = num_attribs + extra_attribs + cat_one_hot_attribs
sorted(zip(feature_importances, attributes) ,reverse = True)
```

### Evaluando el sistema en el conjunto de test

```python
final_model = grid_search.best_estimator_

X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set["median_house_value"].copy()

X_test_prepared = full_pipeline.transform(X_test)

final_predictions = final_model.predict(X_test_prepared)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
```

```python
final_rmse
```

```python
from scipy import stats

confidence = 0.95
squared_errors = (final_predictions - y_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_errors) -1,
                        loc=squared_errors.mean(),
                        scale = stats.sem(squared_errors)))
```

### SVM
